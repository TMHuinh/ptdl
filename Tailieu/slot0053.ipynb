```python
# chuan hoa du lieu
import pandas as pd

year = [2010,2011,2012,2013,2014,2015]

salaryHCM = [8.5,9.0,10.5,12.0,11.0,10.0]

salaryLA = [5.5,6.0,6.0,5.5,6.0,5.0]

df = pd.DataFrame(list(zip(year, salaryHCM,salaryLA)),
                  columns=['Year', 'HCM','LA'])
df
```

```python
# abs max[-1,1]
max_la = df['LA'].abs().max()
df['macLA'] = df['LA'] / max_la

df
```

```python
max_la = df['HCM'].abs().max()
df['macHCM'] = df['HCM'] / max_la

df
```

```python
from sklearn.preprocessing import MaxAbsScaler
scaler = MaxAbsScaler()
scaled = scaler.fit_transform(df[['HCM','LA']])
scaled_df = pd.DataFrame(scaled, columns=df[['HCM','LA']].columns)
print(scaled_df)
```

```python
# luong cua HCM la 4.5, LA = 5.57 thi gia tri scale la bao nhieu
vSc = scaler.transform([[4.5,5.75]])
vSc
```

```python
vR = scaler.inverse_transform([[0.5,0.8]])
vR
```

```python
# min/max scale [0,1]
min_la = df['LA'].min()
max_la = df['LA'].max()
df['mmcLA'] = (df['LA'] - min_la) / (max_la - min_la)

df
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled = scaler.fit_transform(df[['HCM','LA']])
scaled_df = pd.DataFrame(scaled, columns=df[['HCM','LA']].columns)

print(scaled_df)
```

```python
vSc = scaler.transform([[11.0,4.5]])
vSc
```

```python
# StandardScaler 
# Z-Score = x-TB/do lech chuan
mean_la = df['LA'].mean()
std_la = df['LA'].std()
df['zLA'] = (df['LA'] - mean_la) / std_la

df
```

```python
mean_la = df['HCM'].mean()
std_HCM = df['HCM'].std()
df['zHCM'] = (df['HCM'] - mean_la) / std_la

df
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled = scaler.fit_transform(df[['HCM','LA']])

scaled_df = pd.DataFrame(scaled, columns=df[['HCM','LA']].columns)

print(scaled_df)
```

```python
# Transform Function: 
def funcX(x):
    return x*x + 1

df['TransformX'] = funcX(df[["LA"]])

df
```
